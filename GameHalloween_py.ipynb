```python
import random
import time
import os

# -----------------------------
# FunÃ§Ãµes de utilidade
# -----------------------------
def limpar_tela():
    os.system('cls' if os.name == 'nt' else 'clear')

def print_lento(texto, atraso=0.03):
    for letra in texto:
        print(letra, end='', flush=True)
        time.sleep(atraso)
    print()

def linha():
    print("ğŸ•¸ï¸" * 40)

# -----------------------------
# Jogo 1 - Forca Assombrada
# -----------------------------
def forca_assombrada():
    limpar_tela()
    linha()
    print("ğŸƒ  FORCA ASSOMBRADA  ğŸƒ")
    linha()
    print("\nAdivinhe a palavra antes que o monstro te encontre...\n")

    palavras = {
        "fantasma": "Assombra castelos antigos...",
        "bruxa": "Adora voar em noites de lua cheia.",
        "abobora": "SÃ­mbolo clÃ¡ssico do Halloween.",
        "zumbi": "Voltou dos mortos e anda bem devagar.",
        "aranha": "Costuma morar nos cantos escuros.",
        "caveira": "Representa a morte... e o mistÃ©rio."
    }

    palavra, dica = random.choice(list(palavras.items()))
    letras_descobertas = ["_" for _ in palavra]
    tentativas = 6
    usadas = set()

    print(f"Dica: {dica}\n")

    def mostrar_forca(erros):
        estagios = [
            "   -----\n   |   |\n   |   O\n   |  /|\\\n   |  / \\\n  _|_",
            "   -----\n   |   |\n   |   O\n   |  /|\\\n   |  /\n  _|_",
            "   -----\n   |   |\n   |   O\n   |  /|\\\n   |\n  _|_",
            "   -----\n   |   |\n   |   O\n   |  /|\n   |\n  _|_",
            "   -----\n   |   |\n   |   O\n   |   |\n   |\n  _|_",
            "   -----\n   |   |\n   |   O\n   |\n   |\n  _|_",
            "   -----\n   |   |\n   |\n   |\n   |\n  _|_"
        ]
        print(estagios[6 - erros])

    while tentativas > 0 and "_" in letras_descobertas:
        print("Palavra:", " ".join(letras_descobertas))
        print("Letras usadas:", " ".join(sorted(usadas)) or "(nenhuma)")
        mostrar_forca(tentativas)
        palpite = input("\nDigite uma letra: ").lower().strip()

        if not palpite.isalpha() or len(palpite) != 1:
            print("Digite apenas uma letra vÃ¡lida!")
            continue
        if palpite in usadas:
            print("VocÃª jÃ¡ tentou essa letra.")
            continue

        usadas.add(palpite)
        limpar_tela()
        if palpite in palavra:
            for i, c in enumerate(palavra):
                if c == palpite:
                    letras_descobertas[i] = palpite
            print("ğŸƒ Boa! VocÃª acertou uma letra!")
        else:
            tentativas -= 1
            print("ğŸ‘» Letra errada! O monstro estÃ¡ chegando...")
        print()
        time.sleep(1)

    if "_" not in letras_descobertas:
        print(f"\nâœ¨ VocÃª escapou! A palavra era **{palavra.upper()}**! âœ¨")
    else:
        mostrar_forca(0)
        print(f"\nğŸ’€ VocÃª perdeu! A palavra era **{palavra.upper()}**... O monstro te pegou!")

    input("\nPressione Enter para voltar ao menu...")

# -----------------------------
# Jogo 2 - Adivinha dos Doces
# -----------------------------
def adivinha_doces():
    limpar_tela()
    linha()
    print("ğŸ¬  JOÃƒO E MARIA  ğŸ¬")
    linha()
    print("\nUma bruxa escondeu doces em um caldeirÃ£o encantado...\n")

    nivel = input("Escolha a dificuldade (fÃ¡cil / mÃ©dio / difÃ­cil): ").lower()
    if nivel == "facil":
        limite, tentativas = 20, 6
    elif nivel == "medio":
        limite, tentativas = 40, 5
    else:
        limite, tentativas = 80, 4

    numero_secreto = random.randint(1, limite)
    print(f"\nA bruxa cochicha: 'HÃ¡ entre 1 e {limite} doces...'")

    while tentativas > 0:
        try:
            palpite = int(input(f"\nVocÃª tem {tentativas} tentativas. Adivinhe: "))
        except ValueError:
            print("Digite um nÃºmero vÃ¡lido!")
            continue

        if palpite == numero_secreto:
            print("\nâœ¨ CORRETO! VocÃª ganhou todos os doces! ğŸ­âœ¨")
            break
        elif palpite < numero_secreto:
            print("ğŸ”® A bruxa ri: 'Mais... bem mais doces que isso!'")
        else:
            print("ğŸ’€ Ela sussurra: 'Menos... vocÃª exagerou!'")

        tentativas -= 1
        time.sleep(1)

    if tentativas == 0 and palpite != numero_secreto:
        print(f"\nA poÃ§Ã£o evapora... o nÃºmero era {numero_secreto}! Nenhum doce pra vocÃª! ğŸ•¸ï¸")

    input("\nPressione Enter para voltar ao menu...")

# -----------------------------
# Jogo 3 - A Casa do Fantasma
# -----------------------------
def aventura_fantasma():
    limpar_tela()
    linha()
    print("ğŸšï¸  FANTASMA LOUCO ğŸ•¯ï¸")
    linha()
    print()
    print_lento("VocÃª e seus amigos entraram em uma velha mansÃ£o para caÃ§ar doces...")
    print_lento("Mas as portas se fecharam sozinhas!")
    print_lento("Agora, vocÃª precisa encontrar a saÃ­da antes que o fantasma apareÃ§a...\n")

    escolha = input("HÃ¡ trÃªs portas: esquerda, centro e direita. Qual vocÃª escolhe? ").lower()
    if escolha == "esquerda":
        print_lento("\nVocÃª entra em uma biblioteca antiga.")
        acao = input("Quer acender uma vela ou sair? (acender/sair): ").lower()
        if acao == "acender":
            print_lento("A vela acende e revela uma passagem secreta!")
            print_lento("VocÃª escapa pela passagem e vence! ğŸ‰")
        else:
            print_lento("Um livro cai... um espÃ­rito sai dele e te persegue! ğŸ’€ Fim!")
    elif escolha == "centro":
        print_lento("\nVocÃª sobe as escadas que rangem...")
        acao = input("Entrar no quarto iluminado ou no escuro? (iluminado/escuro): ").lower()
        if acao == "iluminado":
            print_lento("VocÃª encontra doces e um bilhete: 'VocÃª estÃ¡ salvo!' ğŸ¬ VitÃ³ria!")
        else:
            print_lento("Um espelho se quebra sozinho... e algo sai dele! ğŸ‘» Fim!")
    elif escolha == "direita":
        print_lento("\nVocÃª entra na cozinha. Um caldeirÃ£o borbulha no fogÃ£o...")
        acao = input("Olhar dentro ou fugir? (olhar/fugir): ").lower()
        if acao == "olhar":
            print_lento("Um braÃ§o sai do caldeirÃ£o e te puxa! ğŸ’€ Fim!")
        else:
            print_lento("VocÃª foge pelos fundos e escapa! ğŸƒ VitÃ³ria!")
    else:
        print_lento("VocÃª hesitou demais... o fantasma apareceu! ğŸ‘» Fim!")

    input("\nPressione Enter para voltar ao menu...")

# -----------------------------
# Menu principal
# -----------------------------
def menu():
    while True:
        limpar_tela()
        linha()
        print("ğŸ®  JOGUINHOS DE HALLOWEEN ğŸ®")
        linha()
        print("\n1 - FORCA ASSOMBRADA ")
        print("2 - JOÃƒO E MARIA ")
        print("3 - FANTASMA LOUCO ")
        print("0 - Sair")
        linha()
        opcao = input("\nEscolha um jogo: ")

        if opcao == "1":
            forca_assombrada()
        elif opcao == "2":
            adivinha_doces()
        elif opcao == "3":
            aventura_fantasma()
        elif opcao == "0":
            limpar_tela()
            print("ğŸ‘» Obrigado por jogar! Feliz Halloween! ğŸƒ")
            break
        else:
            print("OpÃ§Ã£o invÃ¡lida!")
            time.sleep(1)

# -----------------------------
# ExecuÃ§Ã£o principal
# -----------------------------
if __name__ == "__main__":
    menu()
```
