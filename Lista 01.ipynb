```python

# 14. Escreva um programa que solicite ao usuário dois números reais e ao final apresente na
# tela o produto dos dois números informados da seguinte forma:
# "O produto dos numeros N e X corresponde a Y"
num1 = float(input("Digite o primeiro número real: "))
num2 = float(input("Digite o segundo número real: "))
produto = num1 * num2
print("O produto dos números", num1, "e", num2, "corresponde a", produto)
```

```python
# 15. Refazer o programa 14 realizando as quatro operações aritméticas básicas
num1 = float(input("Digite o primeiro número real: "))
num2 = float(input("Digite o segundo número real: "))

soma = num1 + num2
subtracao = num1 - num2
multiplicacao = num1 * num2
divisao = num1 / num2

print("Soma:", soma)
print("Subtração:", subtracao)
print("Multiplicação:", multiplicacao)
print("Divisão:", divisao)

```

```python
# 16. Escreva um programa que solicite o valor fixo do salário de um vendedor, e o total vendido
# (OBS: aqui normalmente se pede calcular comissão, mas não apareceu o enunciado completo.
# Vou assumir comissão de 10% sobre as vendas).
salario_fixo = float(input("Digite o salário fixo do vendedor: "))
total_vendas = float(input("Digite o total vendido: "))

comissao = total_vendas * 0.10
salario_total = salario_fixo + comissao

print("O salário final do vendedor é:", salario_total)
```
